```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def ScoreAnswers(Expected, model_Output):
    client = OpenAI(api_key="<OpenAI key>")

    response = client.embeddings.create(
        input= Expected,
        model="text-embedding-3-small"
    )
    
    v1 = response.data[0].embedding
    response = client.embeddings.create(
        input= model_Output,
        model="text-embedding-3-small"
    )
    v2 = response.data[0].embedding
    embedding1 = np.array(v1).reshape(1, -1)  # Reshape for scikit-learn
    embedding2 = np.array(v2).reshape(1, -1)
    return cosine_similarity(embedding1, embedding2)[0][0]

ScoreAnswers("viva la vida", "React Native UI Library Comparison Tool (RNUILT) - React Native Comparison Tool for UI Libraries and Components.")
```

```python
def calculate_score(row):
    return ScoreAnswers(row['Answer'], row['Model_output'])

df['score'] = df.apply(calculate_score, axis=1)
df['score'].mean()
```
